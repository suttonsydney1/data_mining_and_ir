```python
#install necessary libraries
import requests
```

```python
#import beautiful soup
from bs4 import BeautifulSoup
```

```python
URL = 'http://books.toscrape.com/index.html'
page = requests.get(URL)
print(page.text)
```

```python
#beautifulsoup object
soup = BeautifulSoup(page.content, "html.parser")
```

### Find Book URLs

```python
#find the books listed in the HTML and use soup.find to
#pull the class value corresponding to the tag
soup.find("article", class_="product_pod")
```

```python
#dive deeper into the this by adding div.a 
#then get only the URL contained in the 'href' value
soup.find("article", class_='product_pod').div.a.get('href')
```

```python
#gather ALL product URLs on the main web page
#like above using soup.findAll

main_page_products = [x.div.a.get('href') for x in 
                     soup.findAll('article', class_='product_pod')]

print(str(len(main_page_products)))
print("Fetched Product URLs. One Example: ")
main_page_products[5]
```

```python
#definte getandparse function
def getAndParse(url):
    result = requests.get(url)
    soup  = BeautifulSoup(result.text, 'html.parser')
    return(soup)

#define a function to retrieve book links on any given
#page of the website

def getBooksURLs(url):
    soup = getAndParse(url)
    #remove the index.html part of the base url before returning results
    return(["/".join(url.split("/")[:-1])+ "/" + x.div.a.get('href') for x in 
           soup.findAll("article", class_="product_pod")])
```

### Find Book Categories 

```python
#URL pattern 'catalogue/category/books'

import re

categories_urls = [URL + x.get('href') for x in soup.find_all('a', href=re.compile("catalogue/category/books"))]
categories_urls = categories_urls[1:] #remove 1 because it corresponds to all the books

print(str(len(categories_urls)))
print("Fetched Categories URLs. Examples: ")
categories_urls[:5]
```

### Scrape all books data and product URLs

```python
#get all URLS for all pages
pages_urls = []

new_page = "http://books.toscrape.com/catalogue/page-1.html"
while requests.get(new_page).status_code == 200:
    pages_urls.append(new_page)
    new_page = pages_urls[-1].split("-")[0] + "-" + str(int(pages_urls[-1].split("-")[1].split(".")[0]) + 1) + ".html"
    

print(str(len(pages_urls)) + " fetched URLs")
print("Some examples:")
pages_urls[:5]
```

```python
#get all product URLs
booksURLs = []
for page in pages_urls:
    booksURLs.extend(getBooksURLs(page))
    
print(str(len(booksURLs)))
booksURLs[:5]
```

```python
#get data for each product/book
#information here includes book title, price, availability,image, category, rating

#define lists for these
names = []
prices = []
nb_in_stock = []
img_urls = []
categories = []
ratings = []
```

```python
#scrape the data for every book
for url in booksURLs:
    soup = getAndParse(url)
    
    #product name
    names.append(soup.find("div", class_=re.compile("product_main")).h1.text)
    
    #product price
    prices.append(soup.find("p", class_ = "price_color").text[2:])
    
    #number of availabile products
    nb_in_stock.append(re.sub("[^0-9]", "", soup.find("p", class_ = "instock availability").text)) # get rid of non numerical characters
    
    #image urls
    img_urls.append(url.replace("index.html", "") + soup.find("img").get("src"))
    
    #product category
    categories.append(soup.find("a", href = re.compile("../category/books/")).get("href").split("/")[3])
    
    #ratings
    ratings.append(soup.find("p", class_ = re.compile("star-rating")).get("class")[1])
```

```python
#import all data into a pandas df
import pandas as pd

scraped_data = pd.DataFrame({'name': names, 'price': prices, 'nb_in_stock': nb_in_stock, "url_img": img_urls, "product_category": categories, "rating": ratings})
scraped_data.head()
```

```python
#describe data frame
df = scraped_data

df.describe()
```

```python
#clean up data - remove image url
df.columns
del df['url_img']
```

```python
#change rating items
df["rating"].replace({"One": "1", "Two": "2", "Three":"3", "Four":"4", "Five":"5"}, inplace=True)
df.head()
```

```python
#remove unwanted portions of product category
df['product_category'] = df['product_category'].str.split('_').str[0]
df.head()
```

```python
#group by category to get category stats
cat_df = df.groupby(['product_category']).count()
cat_df = cat_df.rename(columns={'name':'count'})
cat_df = cat_df.drop(['price','nb_in_stock','rating'],axis=1)
```

```python
#sort the dataframe to determine the most popular categories of books
category_count = cat_df.sort_values(by='count', ascending=False)
category_count.head(11)
```

The top 10 categories listed on the website are nonfiction, sequential-art, add-a-comment, fiction, young adult, fantasy, mystery, food-and-drink, and childrens.
